```python
with open('smartprix.html','r',encoding='utf-8') as f:
    html = f.read()
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(html,'lxml')
```

```python
# soup.prettify()
```

```python
containers = soup.find_all('div', class_ = 'sm-product has-tag has-features has-actions')
```

```python
len(containers)
```

```python
import numpy as np
import pandas as pd
```

```python
mobile_name = []
mobile_price = []
mobile_specs_score = []
mobile_sim = []
mobile_processor = []
mobile_space = []
mobile_battery = []
mobile_display = []
mobile_camera = []
mobile_memorycard = []
mobile_version = []
```

```python
for i in containers :
    try:
        mobile_name.append(i.find('h2').text.strip())
    except:
        mobile_name.append(np.nan)

    try:
        mobile_price.append(i.find('span', class_ = 'price').text.strip())
    except:
        mobile_price.append(np.nan)
    
    try:
        mobile_specs_score.append(i.find('b').text.strip())
    except:
        mobile_specs_score.append(np.nan)

    li_list = i.find('ul', class_ = 'sm-feat specs').find_all('li')

    try:
        mobile_sim.append(li_list[0].text.strip())
    except:       
        mobile_sim.append(np.nan)
    
    try:
        mobile_processor.append(li_list[1].text.strip())
    except:       
        mobile_processor.append(np.nan)
        
    try:
        mobile_space.append(li_list[2].text.strip())
    except:       
        mobile_space.append(np.nan)
        
    try:
        mobile_battery.append(li_list[3].text.strip())
    except:       
        mobile_battery.append(np.nan)
        
    try:
        mobile_display.append(li_list[4].text.strip())
    except:       
        mobile_display.append(np.nan)
        
    try:
        mobile_camera.append(li_list[5].text.strip())
    except:       
        mobile_camera.append(np.nan)
        
    try:
        mobile_memorycard.append(li_list[6].text.strip())
    except:       
        mobile_memorycard.append(np.nan)
        
    try:
        mobile_version.append(li_list[7].text.strip())
    except:       
        mobile_version.append(np.nan)
```

```python
mobile_data = pd.DataFrame({
    'mobile_name':mobile_name,
    'mobile_price':mobile_price,
    'mobile_specs_score':mobile_specs_score,
    'mobile_sim':mobile_sim,
    'mobile_processor':mobile_processor,
    'mobile_space':mobile_space,
    'mobile_battery':mobile_battery,
    'mobile_display':mobile_display,
    'mobile_camera':mobile_camera,
    'mobile_memorycard': mobile_memorycard,
    'mobile_version': mobile_version  
})
```
